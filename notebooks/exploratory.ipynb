## 1. Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils import fetch_observacoes
```

## 2. Fetch data from the API

```python
df = fetch_observacoes() 
print("Total records:", len(df))
```

## 3. Basic Statistics

```python
print(df.describe(include="all"))
```

## 4. Species frequency

```python
species_counts = df["especie"].value_counts()
print(species_counts)

species_counts.plot(kind="bar", figsize=(8,4), title="Sepecies Frequency")
plt.show()
```

## 5. Distribution by location

```python
location_species = pd.crosstab(df["localizacao"], df["especie"])
print(location_species)

sns.heatmap(location_species, annot=True, fmt="d", cmap="Blues")
plt.title("Species by Location")
plt.show()
```

## 6. Relationship between environmental conditions and presence

```python
sns.scatterplot(data=df, x="temperatura", y="umidade", hue="especie")
plt.title("Relationship between environmental conditions and species")
plt.show()
```
